<b>Para melhor compreensão da base será divida por estados afins de observar se existe algum estado onde os clientes tem maior remuneração ou melhor distribuição da renda</b>

```python
import seaborn as sns
import pandas as pd
def quantile(x):
    return np.quantile(x, 0.75)

values = df.groupby('Estado')['Renda Mensal'].agg(['mean', 'median', quantile])

g = sns.displot(
    data=df, x='Renda Mensal',
    col='Estado',  
)

axes = g.axes.flatten()

for ax in axes:
    spec = ax.get_title().split(' = ')[1]
    data = values.loc[spec, :]    
    ax.axvline(x=data['mean'], c='green', ls='-', lw=2.5, label='Média: '+str(data['mean']))
    ax.axvline(x=data['median'], c='orange', ls='--', lw=2.5, label='Mediana: '+ str(data['median']))
    ax.axvline(x=data['quantile'], c='red', ls=':', lw=2.5, label='Terceiro quartil: '+ str(data['quantile']))
    ax.legend(loc=0)
```

<b>Em todos os estados observamos que a maioria dos clientes tem uma renda menor do 5000 reais, interessante notar como o terceio quartil de todos os estados em relação a renda é o mesmo valor 3492.0 Então 75% de toda população da base recebe até 3492 reais de renda mensal.</b>

<b>Considerando essa renda, vamos observar como estão os clientes em relação ao endividamente e a renda deles:</b>

```python
import seaborn as sns
import pandas as pd
from matplotlib.pyplot import figure

figure(figsize=(30, 50), dpi=80)

def quantile(x):
    return np.quantile(x, 0.75)

values = df.groupby('Endividamento')['Renda Mensal'].agg(['mean', 'median', quantile])

g = sns.displot(
    data=df, x='Renda Mensal',
    col='Endividamento',  
)

axes = g.axes.flatten()

for ax in axes:
    spec = ax.get_title().split(' = ')[1]
    data = values.loc[spec, :]    
    ax.axvline(x=data['mean'], c='green', ls='-', lw=2.5, label='Média: '+str(data['mean']))
    ax.axvline(x=data['median'], c='orange', ls='--', lw=2.5, label='Mediana: '+ str(data['median']))
    ax.axvline(x=data['quantile'], c='red', ls=':', lw=2.5, label='Terceiro quartil: '+ str(data['quantile']))
    ax.legend(loc=0)
```

<b>Os clientes mais endividados são os que a maior media de renda mensal.</b>

<b>Já conseguimos identificar que os cliente que tem acima de 100% da renda comprometida eles são o grupo com maior renda salárial, agora vamos dividir os grupo de endividamenteos para identificar a quantidade da população</b>

```python
import matplotlib.pyplot as plt
import numpy as np
def bar_plot(df,column, title):
    size = df[column].value_counts()
    labels = df[column].value_counts().keys()
    N = len(list(df[column].value_counts().keys()))
    ind = np.arange(N)
    fig, ax = plt.subplots(figsize = (10,5))
    ax.bar(labels, size, width=0.4)
    
    for index, data in enumerate(size):
        plt.text(x=index , y =data+1 , s=f"{data}" , fontdict=dict(fontsize=20))
    plt.title(title)
    plt.tight_layout()
    plt.show()
bar_plot(df, 'Endividamento', 'Quantidade por Endividamento')
```

<b>Acima de 100% é a segunda maior população presente na base, com 1257 clientes, vamos observar agora em relação a quantidade de restrivos e valores como é o comportamento total da população e depois pelos grupos de endividados</b>

<b>Entre todas as variáveis numéricas vamos tentam identificar agora relaçõs e correlações entre elas, primeiramenet vamos aplicar análises de fatores para melhor entender se existe uma tendência de agrupamento entre as colunas da base. Como a análise fatorial é um método estatístico parametrico é necessário que todas as colunas respeitem a distribuição normal, então vamos normalizar e aplicar o teste, primeiro precisamos identificar quantos factores são encontrados na base</b>

```python
from sklearn.preprocessing import StandardScaler

tmp = df[['Renda Mensal', 'Quantidade de Cheques sem Fundo', 'Quantidade de Restritivos', 'Valor dos Restritivos', 'Quantidade de Protestos', 'Modelo Score 1', 'Modelo Score 2', 'Modelo Score 3', 'Modelo Score 4']]
tmp = tmp[~tmp['Renda Mensal'].isna()]
tmp.head()
scaler = StandardScaler().fit_transform(tmp)
hf_frame_std = pd.DataFrame(scaler, columns=tmp.columns)

from factor_analyzer.factor_analyzer import  FactorAnalyzer
import matplotlib.pyplot as plt
fa = FactorAnalyzer()
fa.fit(hf_frame_std, 25)
# Check Eigenvalues
ev, v = fa.get_eigenvalues()
plt.scatter(range(1, hf_frame_std.shape[1]+1),ev)
plt.plot(range(1, hf_frame_std.shape[1]+1),ev)
plt.title('Scree Plot')
plt.xlabel('Factors')
plt.ylabel('Eigenvalue')
plt.grid()
plt.show()
```

<b>O ponto de inflexição que melhor descreve a quantidade de fatores é o de número 3 então iremos trabalhar com esse valor<b>

```python
fa = FactorAnalyzer(n_factors=3)

fa.fit(tmp)
factors = pd.DataFrame(fa.loadings_, columns=['Factor1', 'Factor2', 'Factor3'])
factors = factors.set_index(tmp.columns)
def highlight_max_(s):
    return ['background-color: tomato' if v > 0.49 or v < -0.49 else '' for v in s]

factors.style.apply(highlight_max_)
```

<b>Entre os fatores, vamos um que é caracterizado pelos modelos de scores de 1 até o 3, outro que é basicamente a quantidade de restritivo e um ultimo que é o modelo de score 4, era esperado que existisse mais colunas repesentando os fatores, porém conseguimos ver que o fator 1 agrupa as informações dos 3 primeiros modelos de score, então vamos ver se existe alguma correlação entre eles</b>

# Modelagem

<b>Iremos agora fazer o modelo de predição, a variavel escolhida foi a Modelo de Score 4, pois seguindo a análise entre os modelos de score</b>

<b>Iremos primeiro transformar as variáveis categoricas em dummies e usaremos regularização com Ridge para para nas variaveis numéricas que tem grandes ranges não representarem bias para o modelo</b>

```python
tmp = df[['Safra', 'Estado', 'Endividamento', 'Performance 30D3M EVER', 'Performance 60D6M EVER']]
tmp = pd.get_dummies(tmp)
tmp1 = df.drop(['Safra', 'Estado', 'Endividamento', 'Performance 30D3M EVER', 'Performance 60D6M EVER', 'Performance 90D12M EVER', 'Performance 90D9M EVER', 'Performance 60D9M EVER'], inplace=False, axis=1)
```

```python
tmp1
```

```python
data = pd.concat([tmp, tmp1],  axis=1)
```

```python
y = data['Modelo Score 4']
X = data[data.columns[:-1]]
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
import warnings
warnings.filterwarnings('ignore')

rf = RandomForestRegressor(random_state=42)
randomforest_params ={'n_estimators':[10,20, 50, 100, None], 'max_depth':[4, 6, 10, 15, 20, 50, None]}
clf_rf = GridSearchCV(rf, randomforest_params, cv=5, n_jobs=12)
clf_rf.fit(X, y)
clf_rf.best_estimator_
```

```python
from sklearn.linear_model import Ridge

ridge = Ridge(random_state=42)
ridge_params = {'alpha':[1,10,100,1000,0.1, 0.01, 0.001, 0, 0.0001]}
clf_ridge = GridSearchCV(ridge, ridge_params, cv=5, n_jobs=12)
clf_ridge.fit(X, y)
clf_ridge.best_estimator_
```

```python
from sklearn.linear_model import Lasso
ridge = Lasso(random_state=42)
lasso_params = {'alpha':[1,10,100,1000,0.1, 0.01, 0.001, 0, 0.0001]}
clf_lasso = GridSearchCV(ridge, lasso_params, cv=5, n_jobs=12)
clf_lasso.fit(X, y)
clf_lasso.best_estimator_
```
